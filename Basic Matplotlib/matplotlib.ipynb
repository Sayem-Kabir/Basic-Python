```python
import matplotlib.pyplot as plt
import pandas as pd
import  numpy as np
```

### Line Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.figure(figsize=(4,2.5), dpi=100)
#plt.plot(x,y, label = '2x', color = 'red', linewidth=2, linestyle = '--', marker = '.',markersize=10, markeredgecolor = 'black')
### Short Hand Notation
# fmt  = '[color][marker][line]'
plt.plot (x,y,'r.--',label='2x') # matplotlib documentation

## 2nd line

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2,'g^-',label='x^2') 
plt.plot(x2[5:], x2[5:]**2,'g--')

plt.title("First Graph", fontdict={'fontname': 'Comic Sans MS','fontsize':20})

plt.xlabel("x_Axis")
plt.ylabel("y_Axis")

plt.xticks([0,1,2,3]) # To remove decimel points
plt.yticks([0,2,4,6,8]) # To remove decimel points

plt.legend()
plt.savefig('mygraph.png', dpi=300) # we need higher dpi but its hard to work with higher dpi, so when we save it we must return for higher dpi
plt.show()
```

### Real Life Problem - GAS Problem

```python
gas = pd.read_csv('gas_prices.csv')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

# plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
# plt.plot(gas.Year, gas.Canada, 'r.-',label='Canada')
# plt.plot(gas.Year, gas['South Korea'], 'g.-',label='South Korea') # for using multiple words
# plt.plot(gas.Year, gas.Australia, 'y.-',label='Australia')

# Another Way to plot many values!
countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.', label = ''+country)

# Print for all countries 
# for country in gas:
#     if country != 'Year':
#         plt.plot(gas.Year, gas[country], marker='.', label = ''+country)

plt.xticks(gas.Year[::3].tolist()+[2011]) ## ::3  skip 3 years  ## .tolist()+[2011] for adding extra values

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

### FIFA Data Load

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

##### Histograms

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall,bins=bins,color='blue')
plt.xticks(bins)


plt.ylabel("Number of Players")
plt.xlabel("Skill Level")
plt.title("Distribution of player skills in FIFA 2018")
plt.show()
```

##### Pie Chart 1

```python
left_foot = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right_foot = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
plt.figure(figsize=(8,5))

labels = ['left','right']
colors=['red','blue']

plt.pie([left_foot,right_foot], labels = labels, colors=colors, autopct='%.2f %%')
plt.title("Foot Preference of FIFA Players")
plt.show()
```

##### Pie Chart 2

```python
plt.figure(figsize=(10,8), dpi=100)

plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4) ## Seperate pies

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

##### Box and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
