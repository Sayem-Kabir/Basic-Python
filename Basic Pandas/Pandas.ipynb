# Value Set

## loc function

```python
newdf.head()
```

```python
newdf.loc[0,0] = 100  #[row,column]
```

```python
newdf.head()
```

```python
newdf.loc[[1,2],[3,4]]  # returns a new data frame, dont change the real data frame
```

```python
newdf.loc[[1,2],:]  # returns all the columns and specific rows
```

```python
newdf.loc[:, [3,4]] # returns all the rows and specific columns
```

### Complex Functions

```python
newdf.loc[(newdf[0]<0.5) & (newdf[3]>0.3)]
```
